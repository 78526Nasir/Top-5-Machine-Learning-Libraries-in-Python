### Stop Words 
__Stop words__ are the commonly used words that doesn't have any impact in analysis. For example: i, a, the, an etc.

```python
# Filter Stop Words
from nltk.corpus import stopwords
```

```python
# list of stop words in english
stop_words = stopwords.words("english")
```

```python
stop_words
```

```python
paragraph = """My name is Nasir Islam Sujan. I am a Student. \
Currently pursuing my BSc. in Software Engineering at Daffodil International University."""
```

```python
paragraph
```

```python
words_token = word_tokenize(paragraph)
words_token
```

```python
filter_words = [word for word in words_token if word not in stop_words]
```

```python
filter_words
```

### Stemming
The process of removing word affixes from the end of words.

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()

words = ["Program", "Programming", "Programmer", "Programmed"]

for word in words:
    print(ps.stem(word))
```

### Parts of Speech Tagging

```python
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
text_2005 = state_union.raw("2005-GWBush.txt")
text_2006 = state_union.raw("2006-GWBush.txt")
```

```python
text_2005
```

```python
cs_tokenizer = PunktSentenceTokenizer(text_2005)
```

```python
cs_tokenizer
```

```python
tokenized = cs_tokenizer.tokenize(text_2006)
```

```python
tokenized
```

```python
def preprocess_text():
        for token in tokenized:
            words = nltk.word_tokenize(token)
            tagged = nltk.pos_tag(words)
            
            print(tagged)
```

```python
preprocess_text()
```

# Scatter Plot using matplotlib

```python
N = 50 # number of random values flag

x = np.random.rand(N) # gives 50 random values and store them into x
y = np.random.rand(N) # gives 50 random values and store them into y

colors = np.random.rand(50) # random colors

area = (30 * np.random.rand(N))**2  # scatter random area

plt.scatter(x, y, label="scatter plot", s = area, c=colors, alpha=0.5) # alpha for scatter opacity, `alpha = 1` means solid color

plt.legend()

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot example")

plt.show()
```

### Increase tha fig size

### Pie Chart

```python
subject_names = ["Math", "Bangla" , "English", "Physics", "Chemistry"]
subject_marks = [90, 70, 33, 68, 47]

colors = ["#ffd50544", "#7952b399", "#ff222244", "#007bff44", "#262c3a44"]

plt.pie(subject_marks, labels = subject_names, autopct='%1.1f%%', startangle=90, colors=colors)
plt.axis('equal') 
plt.show()
```

```python
# increase fontsize and figsize
plt.figure(figsize=(6,6))
subject_names = ["Math", "Bangla" , "English", "Physics", "Chemistry"]
subject_marks = [90, 70, 33, 68, 47]

colors = ["#ffd50544", "#7952b399", "#ff222244", "#007bff44", "#262c3a44"]

plt.pie(subject_marks, labels = subject_names, autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 14})
plt.axis('equal') 
plt.show()
```

### 3D surface (color map)

```python
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter



fig = plt.figure(figsize=(15, 10))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap="RdYlBu", linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
